```python
import pandas as pd
import numpy as np

df = pd.read_csv("./word2vec_wrangling.csv")
kinds_of_exercise = df["exercise_name"].to_list()
```

```python
exercise_name = "ë°”ì°¨íƒ€"
file_name = "#" + exercise_name + "_sum.txt"
file_1 = "/Users/noopy/FitCuration/" + file_name
text = open(file_1, 'r', -1, "UTF-8", errors="ignore").read()
text[:1000]
```

```python
# -*- coding: utf-8 -*-
import re
def preprocessing_hangul(text):
    # ê°œí–‰ë¬¸ì ì œê±°
    hangul = re.compile('[^ ã„±-ã…£ê°€-í£]+')
    result = hangul.sub('', text)
    return result

clean_text = preprocessing_hangul(text)
clean_text[:1000]
```

```python
# MeCabìœ¼ë¡œ ëª…ì‚¬ ë½‘ì•„ë‚´ê¸°
from konlpy.tag import Mecab
mecab = Mecab()
noun_list_mecab = mecab.nouns(clean_text)

# stopwards preprocessing
stopwords_mecab = ['ìˆ˜','í€„ë¦¬í‹°','ë„ì‹œ','ë¶„','ì „ë¬¸','ìŠ¤íƒ€','ë…„','ì›',\
                   'ì›”','í™”','ìˆ˜','ëª©','ê¸ˆ','ì‹œ','ì•¤','ì¼','ê·¸ë¨','ë¬¸'] 

clean_noun_list_mecab = []

for n in noun_list_mecab:
    if n not in stopwords_mecab:
        clean_noun_list_mecab.append(n)

# get top 100 most common nouns
from collections import Counter
nouns_mecab = Counter(clean_noun_list_mecab)

tags_mecab = nouns_mecab.most_common(100)
tags_mecab
```

```python
# Okt(Twitter ì—…ë°ì´íŠ¸ ë²„ì „)ìœ¼ë¡œ ì„œìˆ ì–´ ë½‘ì•„ë‚´ê¸°
# Reference: https://m.blog.naver.com/imsam77/221260229647
# 1. ì´ì „ í¬ìŠ¤íŠ¸ì—ì„œ í¬ë¡¤ë§í•œ ëŒ“ê¸€íŒŒì¼ì„ ì½ê¸°ì „ìš©ìœ¼ë¡œ í˜¸ì¶œí•¨
# Trying to get adjectives
# 3. íŠ¸ìœ—í„° íŒ¨í‚¤ì§€ ì•ˆì— konlpy ëª¨ë“ˆí˜¸ì¶œ
from konlpy.tag import Okt
from collections import Counter

twitter = Okt()

# 4. ê° ë¬¸ì¥ë³„ë¡œ í˜•íƒœì†Œ êµ¬ë¶„í•˜ê¸°
morphed_list_okt = twitter.pos(clean_text)
# morphed_list

# ë¶ˆìš©ì–´
stopwords_Twitter = ["ì…ë‹ˆë‹¤","ìˆëŠ”","ìˆìŠµë‹ˆë‹¤","ê°™ì€","ì•ˆë…•í•˜ì„¸ìš”","ê³ ë§ˆì›Œìš”","ìˆì–´ìš”","ìˆê²Œ"\
                     ,"ìˆë„ë¡","ë¶€íƒë“œë¦½ë‹ˆë‹¤","í•˜ëŠ”","í•©ë‹ˆë‹¤","í• ","í•˜ì„¸ìš”","í•˜ê¸°","í•´","ë©ë‹ˆë‹¤","í•˜ì—¬",'ì˜','ëœ','ë˜ê³ ','ë˜ì–´','ë˜ì—ˆìŠµë‹ˆë‹¤',"ì—†ëŠ”","ë“œë¦½ë‹ˆë‹¤"\
                    ,"ë˜ê¸°","í•˜ì‹œëŠ”","í•˜ê³ ","ì•Šì„","ê°™ë‹¤","ì‹¶ë‹¤","ì´ëŸ°","ì €ëŸ°","ê·¸ëŸ°",'ë°”ëë‹ˆë‹¤'\
                    ,"í–ˆìŠµë‹ˆë‹¤","í–ˆë‹¤","í•´ë“œë¦½ë‹ˆë‹¤","í•˜ì‹ ","í•˜ì‹¤","ì•Šê³ ","í•´ìš”","ê°€ëŠ¥í•©ë‹ˆë‹¤","í•˜ê³ ì‹¶ìœ¼ì‹ "\
                    ,"ì•Šìœ¼ë©°","ì£¼ì„¸ìš”","ì˜¤ì„¸ìš”"]

# 5. í˜•ìš©ì‚¬ í’ˆì‚¬ë§Œ ë½‘ì•„ë‚´ê¸°
adj_list_okt = []
for word, tag in morphed_list_okt:
    if tag in ['Adjective','Verb'] and word not in stopwords_Twitter:
        adj_list_okt.append(word)

# 6. ì„ ë³„ëœ í’ˆì‚¬ë³„ ë¹ˆë„ìˆ˜ ê³„ì‚° & ìƒìœ„ ë¹ˆë„ 10ìœ„ ê¹Œì§€ ì¶œë ¥
adj_counts_okt = Counter(adj_list_okt)
common_adj_okt = adj_counts_okt.most_common(100)
common_adj_okt
```
